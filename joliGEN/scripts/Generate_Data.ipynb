```python
import os
import cv2
import time
import shutil
import pandas as pd
import numpy as np
from PIL import Image
from PIL import Image as PilImage
from matplotlib import pyplot as plt

import generate_images
```

```python
def get_data_cut(path_source, path_target, model_in_file):
    cpu = "store_true"
    gpuid = 0
    files_list = sorted(os.listdir(path_source))
    for curr_file in files_list:
        img_in = f"{path_source}/{curr_file}"
        img_out = f"{path_target}/new_{curr_file}"
        generate_images.main(model_in_file, img_in, img_out, cpu, gpuid)
```

```python
def get_df_data(path_target, label, sens):
    files_list = sorted(os.listdir(path_target))
    elements = []
    for curr_file in files_list:
        elements.append((curr_file, label, sens))
    df = pd.DataFrame(elements, columns=["image_id", "Male", "Sens"])
    return df
```

```python
model_in_file = "/home/vuong.nguyen/vuong/augmentare/joliGEN/checkpoints/male_gray_2_female_gray_cut/latest_net_G_A.pth"
path_source = "/home/vuong.nguyen/vuong/augmentare/joliGEN/dataset/male_gray_2_female_gray/trainA"
path_target = "/home/vuong.nguyen/vuong/augmentare/Bias Celeba/new_dataset/CUT_face/female_gray"
get_data_cut(path_source, path_target, model_in_file)
```

```python
df = get_df_data(path_target, 0, 4)
```
