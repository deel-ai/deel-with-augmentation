# Explainable AI

## ResNet18

### Highway

```python
hw_resnet18_path = "/home/vuong.nguyen/vuong/augmentare/Bias in EuroSAT/X_HW/Origin/ResNet18"
```

```python
fig = plt.figure(figsize=(8,12))
rows = ['Origin', 'GradCAM', 'GradCAMPlus', 'LayerCAM', 'ScoreCAM', 'Ig', 'Guidedbp', 'Shap']
for num in range(1,9):
    if num == 1:
        ax = fig.add_subplot(8, 1, num)
        img = PilImage.open(f"/home/vuong.nguyen/vuong/augmentare/Bias in EuroSAT/X_HW/Origin/hw.png")
        ax.imshow(img, aspect='auto')
        ax.set_ylabel(rows[num-1], rotation=0, fontsize=15, labelpad=50)
        ax.yaxis.label.set_color('black')
        ax.set_xticks([])
        ax.set_yticks([])

    elif num > 1:
        ax = fig.add_subplot(8, 1, num)
        img = PilImage.open(f"{hw_resnet18_path}/Baseline/res_{num-1}.png")
        ax.imshow(img, aspect='auto')
        ax.set_ylabel(rows[num-1], rotation=0, fontsize=15, labelpad=50)
        ax.yaxis.label.set_color('black')
        ax.set_xticks([])
        ax.set_yticks([])

fig.tight_layout()
plt.show()
```
